```python
import spacy
```

```python
nlp = spacy.load("en_core_web_lg") 
```

```python
'and' in nlp.Defaults.stop_words
```

```python
'coffee' in nlp.Defaults.stop_words
```

```python
# add 'coffee' to stop words
nlp.Defaults.stop_words = nlp.Defaults.stop_words.union({'coffee'})
```

```python
'coffee' in nlp.Defaults.stop_words
```

```python
doc = nlp('coffee and tea')
for token in doc:
    print(f'"{token.lemma_}"')
    print(f'is_stop: {token.is_stop}')
    print(f'is in set: {token.lemma_ in nlp.Defaults.stop_words}\n')
```

```python
nlp.Defaults.stop_words.add("espresso")
```

```python
'espresso' in nlp.Defaults.stop_words
```

```python
doc = nlp('espresso and tea')
for token in doc:
    print(f'"{token.lemma_}"')
    print(f'is_stop: {token.is_stop}')
    print(f'is in set: {token.lemma_ in nlp.Defaults.stop_words}\n')
```
